#### Part Two: Logic Puzzles

```python
def Puzzle():
    Names = 'Ana Glenda Jessie Lara Rose'.split()
    Colors = 'blue green red white yellow'.split()
    Ages = '24 26 28 30 32'.split()
    Professions = 'biologist locksmith judge nurse singer'.split()
    Countries = 'China Egypt Italy Mexico Peru'.split()
    Durations = '5d 10d 15d 20d 25d'.split()
    variables = Names + Colors + Ages + Professions + Countries + Durations
    domains = {}
    for var in variables:
        domains[var] = list(range(1, 6))
    domains['singer'] = [3]
    domains['judge'] = [1]
    domains['15d'] = [2]
    neighbors = parse_neighbors("""Peru: 20d; 20d: blue;
                Ana: biologist; 32: Egypt; Peru: white; green: Glenda; 
                30: white; white: blue; 24: Mexico; white: 24; 20d: Italy; 
                25d: red; 20d: nurse; Lara: locksmith; locksmith: blue;
                Rose: green; 5d: 32; 5d: 28; 30: blue; blue: yellow""", variables)
    for type in [Names, Colors, Ages, Professions, Countries, Durations]:
        for A in type:
            for B in type:
                if A != B:
                    if B not in neighbors[A]:
                        neighbors[A].append(B)
                    if A not in neighbors[B]:
                        neighbors[B].append(A)

    def puzzle_constraint(A, a, B, b, recurse=0):
        same = (a == b)
        next_to = abs(a - b) == 1
        near = abs(a - b) == 2
        if A == 'Peru' and B == '20d':
            return near
        if A == '20d' and B == 'blue':
            return near
        if A == 'Ana' and B == 'biologist':
            return next_to
        if A == '32' and B == 'Egypt':
            return same
        if A == 'Peru' and B == 'white':
            return next_to
        if A == 'green' and B == 'Glenda':
            return near
        if A == '30' and B == 'white':
            return near
        if A == 'white' and B == 'blue':
            return near
        if A == '24' and B == 'Mexico':
            return same
        if A == 'white' and B == '24':
            return near
        if A == '20d' and B == 'Italy':
            return next_to
        if A == '25d' and B == 'red':
            return same
        if A == '20d' and B == 'nurse':
            return next_to
        if A == 'Lara' and B == 'locksmith':
            return near
        if A == 'locksmith' and B == 'blue':
            return near
        if A == 'Rose' and B == 'green':
            return same
        if A == '5d' and B == '32':
            return next_to
        if A == '5d' and B == '28':
            return same
        if A == '30' and B == 'blue':
            return near
        if A == 'blue' and B == 'yellow':
            return near
        if recurse == 0:
            return puzzle_constraint(B, b, A, a, 1)
        if ((A in Names and B in Names) or
                (A in Colors and B in Colors) or
                (A in Ages and B in Ages) or
                (A in Professions and B in Professions) or
                (A in Countries and B in Countries) or
                (A in Durations and B in Durations)):
            return not same
        raise Exception('error')
    return CSP(variables, domains, neighbors, puzzle_constraint)
```

```python
def solve_puzzle(algorithm=backtracking_search, **args):
    p = Puzzle()
    ans = algorithm(p, **args)
    for h in range(1, 7):
        print('Traveler', h, end=' ')
        for (var, val) in ans.items():
            if val == h:
                print(var, end=' ')
        print()
    return ans
```

```python
result = solve_puzzle()
```

```python
result = solve_puzzle(algorithm=min_conflicts)
```

```python
def class_scheduling():
    Classes = 'CH100 GH110 GH112 AMH220 AMH230 GH241 CH245 F271-1 F271-2 F275-1 K275-2 T280 CH301 AMH320'.split()
    Schedule = '215/MWF9 215/MWF1020 215/MWF1140 215/MWF1 215/TR9 215/TR1055 215/TR1250 221/MWF9 221/MWF1020 221/MWF1140 221/MWF1 221/TR9 221/TR1055 221/TR1250 222/MWF9 222/MWF1020 222/MWF1140 222/MWF1 222/TR9 222/TR1055 222/TR1250'.split()
    variables = Classes + Schedule
    domains = {}
    for var in variables:
        domains[var] = list(range(1, 22))
    neighbors = parse_neighbors("""GH112: 222; AMH220: 222; AMH230:221; GH241: 222; CH245: 215; F271-1: 215;
                F271-2: 215; F275-1: 215; K275-2: 215; T280: 222; CH301: 221; AMH320:222; 
                AMH220: MWF9; CH: MWF1; F271-1: TR1250; F271-2: TR9; T280: MWF1TR1250; K: 9;
                CH301: GH110; CH100: GH112; CH100: AMH220; CH100: F271; GH112: AMH220; GH112: F271; AMH220: F271""", variables)
    for type in [Classes, Schedule]:
        for A in type:
            for B in type:
                if A != B:
                    if B not in neighbors[A]:
                        neighbors[A].append(B)
                    if A not in neighbors[B]:
                        neighbors[B].append(A)

    def schedule_constraint(A, a, B, b, recurse=0):
        same = (a == b)
        diff = abs(a - b) == 1
        if A == 'GH112' and '222' in B:
            return same
        if A == 'AMH220' and '222' in B:
            return same
        if A == 'AMH230' and '221' in B:
            return same
        if A == 'GH241' and '222' in B:
            return same
        if A == 'CH245' and '215' in B:
            return same
        if A == 'F271-1' and '215' in B:
            return same
        if A == 'F271-2' and '215' in B:
            return same
        if A == 'F275-1' and '215' in B:
            return same
        if A == 'K275-2' and '215' in B:
            return same
        if A == 'T280' and '222' in B:
            return same
        if A == 'CH301' and '221' in B:
            return same
        if A == 'AMH320' and '222' in B:
            return same
        if A == 'AMH220' and 'MWF9' in B:
            return same
        if 'CH' in A and 'MWF1' in B:
            return diff
        if 'F271-1' in A and 'TR' in B and '1250' not in B:
            return same
        if 'F271-2' in A and 'TR' in B and '9' not in B:
            return 
        if A == 'T280' and 'MWF1' in B or 'TR1250' in B:
            return same
        if A == 'K' and '9' in B:
            return same
        if A == 'CH301' and B != 'GH110':
            return diff
        if A == 'CH100' and B == 'GH112':
            return diff
        if A == 'CH100' and B == 'AMH220':
            return diff
        if A == 'CH100' and 'F271' in B:
            return diff
        if A == 'GH112' and B == 'AMH220':
            return diff
        if A == 'GH112' and 'F271' in B:
            return diff
        if A == 'AMH220' and 'F271' in B:
            return diff
        if recurse == 0:
            return schedule_constraint(B, b, A, a, 1)
        if ((A in Classes and B in Classes) or
                (A in Schedule and B in Schedule)):
            return not same
        raise Exception('error')
    return CSP(variables, domains, neighbors, schedule_constraint)
```
